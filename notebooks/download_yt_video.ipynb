```python
#Code to download the youtube video

from pytubefix import YouTube
```

```python
video_path="https://youtu.be/KzYm9Qe1pTs?si=9dpG7ulZ2186KYAt"

yt = YouTube(video_path)
stream = yt.streams.filter(file_extension='mp4', res='1080p').first()
stream.download(filename="SRH-RCB-2024.mp4")
```

```python
video_url = "https://youtube.com/shorts/eh01Dd8H9PA?si=Gc9l1fFwISmlD-qY"
yt = YouTube(video_url)

stream = yt.streams.get_highest_resolution()
stream.download(filename="Test-Video2-SRH-RCB-2024.mp4")
print("✅ Download complete!")
```
